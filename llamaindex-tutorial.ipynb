```python
from llama_index.readers.web import SimpleWebPageReader
documents = SimpleWebPageReader(html_to_text=True).load_data(["http://paulgraham.com/worked.html"])
print(f"Loaded {len(documents)} document(s)")

# show the preview of the document
documents[0].get_content()
print(documents[0].get_content())
```

1. An Index is a data structure that allows us to quickly retrieve relevant context for a user query.

Documents → [Index organizes them] → User asks question → [Index finds answer] → Response

```python
from llama_index.core import VectorStoreIndex

print("creating index")
index = VectorStoreIndex.from_documents(documents)   
print("index created")

#create a query engine
query_engine = index.as_query_engine()

#ask a question
question = "Does paul graham knows artificial intelligence?"
response = query_engine.query(question)
print(response)
```
